A dataframe can be described as multiple different series combined with one another

```python
import numpy as np
import pandas as pd
cars_csv = 'https://raw.githubusercontent.com/AhsenRiaz/Numpy-And_Pandas/main/cars.csv'
```

```python
sales_np = np.random.randint(1000, 5000, (4,3))
print(sales_np, "\n")

names = ["Tom", "John", "Bob", "Jerry"]
print(names, "\n")

months = ["April", "May", "June"]
print(months, "\n")

sales_df = pd.DataFrame(sales_np, index=names, columns=months)
print(sales_df);
```
