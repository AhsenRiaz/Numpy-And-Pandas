We can use pandas methods for accessing or manipulating the data points

```python
import numpy as np
import pandas as pd
```

```python
cars_csv='https://raw.githubusercontent.com/AhsenRiaz/Numpy-And_Pandas/main/cars.csv'

df = pd.read_csv(cars_csv)
df
```

# Apply
we can insert certain new coloumns through this method

```python
def price_after_tax(price):
  return round((price * 1.13)) # a tax of 13 percent
```

```python
price_after_tax(df['price'][7])
```

```python
df['price_after_tax'] =df['price'].apply(price_after_tax) # aply tax to each price and create a new col with price after taxes
df
```

```python
def profit_margin(price, finance):
    if finance == "no":
        if price < 10000:
            return "LOW"
        elif price >= 10000 and price < 25000:
            return "MED"
        else:
            return "HIGH"
    else:
        if price < 7500:
            return "LOW"
        elif price >= 7500 and price < 20000:
            return "MED"
        else:
            return "HIGH"
```

```python
df["profit_margin"] = df[['price', 'finance']].apply(lambda df: profit_margin(df['price'], df['finance']), axis=1)
df
```

```python
# Another method for this
df['profit_margin'] = np.vectorize(profit_margin)(df['price'], df['finance'])
df
```
